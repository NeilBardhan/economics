# Examining the Consumer Price Index

```python
import os
import sys
import json
import requests
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
os.chdir('..')
sys.path.append(os.getcwd() + '/src/')
```

## Reading in the CPI

```python
cpi_df = pd.read_csv(os.getcwd() + '/data/CPI.csv', header='infer')
cpi_df.head()
```

## Filtering the CPI on `date`

```python
cpi_df['date'] =  pd.to_datetime(cpi_df['date'], format='%Y-%m-%d')
cpi_5y = cpi_df[cpi_df['date'] >= '2016-01-01']
```

## Plotting the CPI

```python
plt.figure(figsize=(16, 9), dpi=300)
plt.plot(cpi_5y['date'], cpi_5y['CPIAUCSL'], color='red');
```
